# Multiple Plots

```python
x = np.arange(100)
```

```python
fig, axes = plt.subplots(2, 2)
fig.suptitle("Four Plots")

axes[0, 0].plot(x, np.sin(x))
axes[0, 0].set_title("Sine Wave")

axes[0, 1].plot(x, np.cos(x))
axes[0, 1].set_title("Cosine Wave")

axes[1, 0].plot(x, np.random.random(100))
axes[1, 0].set_title("Random Function")

axes[1, 1].plot(x, np.log(x))
axes[1, 1].set_title("Log Function")
axes[1, 1].set_xlabel("X")
axes[1, 1].set_ylabel("Log(X)")

fig.tight_layout()

plt.savefig("four_plots.png", dpi=300, bbox_inches="tight", pad_inches=0.25)

plt.show() 
```

# 3D Plotting and Animations

```python
def plot_3d (x, y, z):
    ax = plt.axes(projection="3d")
    ax.plot(x, y, z)
    ax.set_title("3D Plot")
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    plt.show() 
```

```python
x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)
```

```python
plot_3d(x, y, z)
```

```python
ax = plt.axes(projection="3d")
ax.scatter(x, y, z)
ax.set_title("3D Plot")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show() 
```

# Surface Plots

```python
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)
```

```python
ax = plt.axes(projection="3d")
ax.plot_surface(X, Y, Z)
ax.set_title("Surface Plot")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show() 
```

```python
ax = plt.axes(projection="3d")
ax.plot_surface(X, Y, Z, cmap="turbo")
ax.set_title("Surface Plot")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
# ax.view_init(azim=45, elev=30)
plt.show() 
```
